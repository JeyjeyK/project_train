```python
import pandas as pd
import csv
```

```python
file_names = ['CARD_SUBWAY_MONTH_2018.csv','CARD_SUBWAY_MONTH_2019.csv']
total = pd.DataFrame()
for file_name in file_names:
    temp = pd.read_csv(file_name,encoding='cp949')
    total = pd.concat([total,temp])
```

```python
file_path = 'CARD_SUBWAY_MONTH_2021.csv' # 컬럼이 달라 다른 방식으로 해줘야함 
f = open(file_path,'rt')
reader = csv.reader(f)

csv_list = []
for l in reader:
    csv_list.append(l)
f.close()

df = pd.DataFrame(csv_list)
df = df.rename(columns=df.iloc[0])
df = df.drop(df.index[0])
df = df.drop([None], axis = 1)
total = pd.concat([total,df])
```
